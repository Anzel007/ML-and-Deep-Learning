<h3> 1. K Means Clustering</h3>
- grouping data into groups(eg 2)
- now centroid(central value) is choosed for the clusters
- the average dist is calcualted from centroid to each points in the cluster
- the average value must match the centroid value
-if it doesnt match then the calcualted average value is taken as new centroid
-then average distance is calculated similary 
- now now the calculated average value and the centroid is compared .....
then the process is repeated
