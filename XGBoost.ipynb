```python
import xgboost as xgb

##defining data_matrix
data_dmatrix= xgb.DMatrix(data=X,label=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from xgboost import XGBClassifier

#DECLAREE PARAMETERS  
params={
      'objective':'binary:logistic',
       'max_depth':4,
      'alpha':10,
     'learning_rate':1.0,
     'n_estimators':100
        
}

#instantiate the classifier

xgb_clf=XGBClassifier(**params)
xgb_clf.fit(X_train, y_train)


```

```python
#alternatively view the parameters of the xgb trained model
print(xgb_clf)
```

```python
##make prediction
y_pred = xgb_clf.predict(X_test)
```

```python
###check accuracy score
#k accuracy score
from sklearn.metrics import accuracy_score
print('XGBoost model acuracy score:{0:0.4f}',format(accuracy_score(y_test,y_pred)))

```

```python
from sklearn import svm, datasets
iris = datasets.load_iris()
```

```python
df= pd.DataFrame(iris.data,columns=iris.feature_names)
df['flower']=iris.target
df['flower']=df['flower'].apply(lambda x:iris.target_names[x])
df[47:150]

```

```python
####        USE TRAIN TEST SPLIT AND MANUALLY TUNE PARAMETERS BY TRIAL AND ERROR
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
X_train, X_test, y_train,y_test= train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
model=svm.SVC(kernel='rbf', C=30,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
########### USE K FOLD CROSS VALIDATION ########
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),iris.data,iris.target)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target)
```

```python
kernels=['rbf','linear']
C=[1,10,20]
avg_scores={}
for kval in kernels:
  for cval in C:
    cv_scores=cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),iris.data,iris.target)
    avg_scores[kval+ '_' +str(cval)]=np.average(cv_scores)
    
avg_scores
```

```python
#from above data we see that rbf withb C=1 or 10 or linear with C=1 will give best results
```

```python
##### HYPER PARAMETER TUNING #########
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
}, cv=5,return_train_score=False)
clf.fit(iris.data, iris.target)
clf.cv_results_

```
